```python
import matplotlib.pyplot as plt
from numpy import arange
from numpy import meshgrid
from numpy import exp as npexp
from math import exp as mathexp
from mft7 import Newton_method
```

### Дано: система
### $e^{y-\frac{x}{10}}-xy = 1.4$
### $x^2+2y^2=4$
### Строим графики для определения начального приближения решений системы

```python
delta = 0.025
xrange = arange(-3.0, 3.0, delta)
yrange = arange(-3.0, 3.0, delta)
X, Y = meshgrid(xrange,yrange)
F = npexp(Y-X/10)-X*Y-1.4
G = X**2+2*Y**2-4
plt.contour(X, Y, F, [0],colors = "purple")
plt.contour(X, Y, G, [0],colors = 'c')
plt.grid()
plt.show()
```

```python
X0 = [-2.0, 0.9, 1.3, 1.8]
Y0 = [0.0, -1.3, 1.2, -0.5]
plt.contour(X, Y, F, [0],colors = "purple")
plt.contour(X, Y, G, [0],colors = 'c')
for i in range(4):
    plt.plot(X0[i],Y0[i],'ro') 
plt.grid()
plt.show()
```

### Описание функций $f$ и $g$ и их частных производных.
### Также объявление максимального числа итераций и точности

```python
f   = lambda x,y: mathexp(y-x/10)-x*y-1.4
g   = lambda x,y: x**2 + 2*y**2-4
dfx = lambda x,y: -mathexp(y-x/10)/10-y
dfy = lambda x,y: mathexp(y-x/10)-x
dgx = lambda x,y: 2*x
dgy = lambda x,y: 4*y
df = {'x':dfx, 'y':dfy}
dg = {'x':dgx, 'y':dgy}

kmax = 10
eps = 10**(-5)
```

## 1) Уточняем решение $A0$

```python
dataframe, solution = Newton_method(f, g, df, dg, X0[0], Y0[0], kmax, eps)
print("Уточнённое решение A0:{0}".format(solution))
dataframe
```

## 2) Уточняем решение $A1$

```python
dataframe, solution = Newton_method(f, g, df, dg, X0[1], Y0[1], kmax, eps)
print("Уточнённое решение A1:{0}".format(solution))
dataframe
```

## 3) Уточняем решение $A2$

```python
dataframe, solution = Newton_method(f, g, df, dg, X0[2], Y0[2], kmax, eps)
print("Уточнённое решение A2:{0}".format(solution))
dataframe
```
