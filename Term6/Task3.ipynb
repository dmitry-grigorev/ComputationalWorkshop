```python
import numpy as np
import random
import time
import Task3.moduIe_for_task3.mft3 as mft3
random.seed(231)
```

## Генерируем "вершины" выпуклого многоугольника

```python
dots = mft3.get_vertices_in_circle(1000)
mft3.plot_dots(dots)
```

## Строим выпуклое покрытие точек, то есть многоугольник

```python
cov = mft3.build_coverance(dots)
```

```python
mft3.plot_coverance(dots, cov)
```

### Как считается следующая вершина:
### Каждый раз считается угол между вектором, который направлен от новой вершины оболочки к старой и от новой до искомой следующей
### При этом выбирается та из вершин, для которой синус половинного угла между этими векторами наибольший (так и угол наибольший), то же что 1-$\cos \phi$ больше

## Триангуляризуем многоугольник

```python
triangles = mft3.triangularize(cov)
```

```python
mft3.plot_triangles(cov, triangles)
```

## Наконец, моделируем равномерное распределение в многоугольнике

```python
generated = mft3.polygon_uniform(triangles, np.power(10,3))
```

```python
mft3.plot_dots_and_triangles(generated.T, triangles, cov)
```

## Увеличим число брошенных точек

```python
start = time.time()
generated = mft3.polygon_uniform(triangles, np.power(10,6))

print(time.time() - start, "seconds elapsed")
```

```python
mft3.plot_dots_and_triangles(generated.T, triangles, cov)
```
